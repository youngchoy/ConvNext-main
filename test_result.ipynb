```python
import matplotlib.pyplot as plt
import pandas as pd

result_dir = './path/to/save_results/log.txt'

df = pd.read_table(result_dir, sep=",")
```

```python
train_lr = list()
train_min_lr = list()
train_loss = list()
test_acc1 = list()
test_acc5 = list()
epoch = list()
n_parameters = list()
test_loss_ema = list()
test_acc1_ema = list()
test_acc5_ema = list()

length = df.shape[0]
for i in range(length):
    train_lr.append(float(df.iloc[i][0].split(":")[1]))
    train_min_lr.append(float(df.iloc[i][1].split(":")[1]))
    train_loss.append(float(df.iloc[i][2].split(":")[1]))
    train_loss.append(float(df.iloc[i][3].split(":")[1]))
    test_acc1.append(float(df.iloc[i][4].split(":")[1]))
    test_acc5.append(float(df.iloc[i][5].split(":")[1]))
    epoch.append(float(df.iloc[i][6].split(":")[1]))
    n_parameters.append(float(df.iloc[i][7].split(":")[1]))
    test_loss_ema.append(float(df.iloc[i][8].split(":")[1]))
    test_acc1_ema.append(float(df.iloc[i][9].split(":")[1]))
    test_acc5_ema.append(float(df.iloc[i][10].split(":")[1]))
    row = df.iloc[i]
print(row)
```
